```python
def sum_alist(x:list):
  """
  function that takes the sum of a list x
  """
  sum = 0
  for element in x:
    print("this is the element to added to the sum", element)
    # sum = sum + element
    sum += element
    print("this is the running sum", sum)
```

```python
# this is a test list to be summed
test_list = [1, 10.0, 1e10, 3.14]
# this sums the list
sum_of_test_list = sum_alist(test_list)
print(sum_of_test_list)
```
